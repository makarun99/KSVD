```python
# Лабораторна робота №1: Аналіз даних
# Студент: Бабич Марко
# Предмет: Комп'ютерні системи візуалізації даних

import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt

iris = px.data.iris()

fig = px.scatter_3d(
    iris,
    x='sepal_length',
    y='sepal_width',
    z='petal_length',
    color='species',
    title='<b style="text-align: center;">3D-розподіл квіток Iris</b>'
)

fig.update_layout(title_x=0.5)  # Вирівнювання по центру
fig.show()
fig = px.histogram(
    iris,
    x='petal_width',
    color='species',
    title='<b style="text-align: center;">Гістограма розподілу petal_width для різних видів Iris</b>'
)

fig.update_layout(title_x=0.5)
fig.show()
mean_data = iris.groupby(['species', 'sepal_width']).mean().reset_index()

fig = px.line(
    mean_data,
    x='sepal_width',
    y='sepal_length',
    color='species',
    title='<b style="text-align: center;">Зміна середнього sepal_length залежно від sepal_width</b>'
)

fig.update_layout(title_x=0.5)
fig.show()
gapminder = px.data.gapminder()
gap_2007 = gapminder[gapminder['year'] == 2007]
top10_gdp = gap_2007.nlargest(10, 'gdpPercap')

fig = px.bar(
    top10_gdp,
    x='country',
    y='gdpPercap',
    color='gdpPercap',
    text='gdpPercap',
    title='<b style="text-align: center;">Топ-10 країн за ВВП у 2007 році</b>'
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(title_x=0.5)
fig.show()
europe_2007 = gapminder[(gapminder['continent'] == 'Europe') & (gapminder['year'] == 2007)]
top5_europe = europe_2007.nlargest(5, 'gdpPercap')

fig = px.pie(
    top5_europe,
    values='gdpPercap',
    names='country',
    title='<b style="text-align: center;">Розподіл ВВП топ-5 країн Європи у 2007 році</b>'
)

fig.update_layout(title_x=0.5)
fig.show()
canada_oceania = gapminder[(gapminder['continent'] == 'Oceania') | (gapminder['country'] == 'Canada')]

fig = px.bar(
    canada_oceania,
    x='year',
    y='pop',
    color='country',
    barmode='group',
    title='<b style="text-align: center;">Динаміка населення Канади та країн Океанії</b>'
)

fig.update_layout(title_x=0.5)
fig.show()
x = np.linspace(0, 5*np.pi, 500)
y = 2 * np.cos(x) * (x - np.pi/4)

plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.title('Графік функції $y = 2 \\cos(x)(x - \\frac{\\pi}{4})$', loc='center')
plt.grid()
plt.show()
x = np.linspace(0, 2*np.pi, 500)
y = 0.5 * np.sin(2*x + np.pi/3)

plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.title('Графік функції $y = \\frac{1}{2} \\sin(2x + \\frac{\\pi}{3})$', loc='center')
plt.grid()
plt.show()
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = 2 * np.tan(0.5*x + np.pi/6)

plt.figure(figsize=(8,5))
plt.plot(x, y)
plt.ylim(-10, 10)  # обмеження для наочності
plt.title('Графік функції $y = 2 \\tan\\left(\\frac{1}{2}x + \\frac{\\pi}{6}\\right)$', loc='center')
plt.grid()
plt.show()
```
