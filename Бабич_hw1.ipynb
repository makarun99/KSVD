```python
# Лабораторна робота №1: Аналіз даних
# Студент: Бабич Марко
# Предмет: Комп'ютерні системи візуалізації даних

import numpy as np

# 1. Створіть одновимірний масив (вектор) з першими 20-ма натуральними числами та виведіть його значення.
vector_20 = np.arange(1, 21)
print("Одновимірний масив з перших 20 натуральних чисел:\n", vector_20)

# 2. Створіть двовимірний масив (матрицю) розміром 4x4, заповніть його одиницями та виведіть його значення.
matrix_4x4 = np.ones((4, 4))
print("\nМатриця 4x4, заповнена одиницями:\n", matrix_4x4)

# 3. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від -1 до 1 та виведіть його значення.
random_5x5 = np.random.uniform(-1, 1, (5, 5))
print("\nМатриця 5x5 з випадковими числами [-1, 1]:\n", random_5x5)

# 4. Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та виведіть його значення.
random_3x3 = np.random.rand(3, 3)
print("\nМатриця 3x3 з випадковими числами [0, 1]:\n", random_3x3)

# 5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від 1 до 20
a = np.random.randint(1, 21, 6)
b = np.random.randint(1, 21, 6)
print("\nМасив A:", a)
print("Масив B:", b)

# Поелементні операції
print("A + B:", a + b)
print("A - B:", a - b)
print("A * B:", a * b)

# 6. Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.
mat1 = np.random.randint(1, 6, (3, 2))
mat2 = np.random.randint(1, 6, (2, 4))
product = np.dot(mat1, mat2)
print("\nМатриця 3x2:\n", mat1)
print("Матриця 2x4:\n", mat2)
print("Результат множення 3x2 * 2x4:\n", product)

# 7. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.
mat_trans = np.random.rand(3, 3)
print("\nОригінальна матриця 3x3:\n", mat_trans)
print("Транспонована матриця:\n", mat_trans.T)

# 8. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.
mat_a = np.random.randint(1, 11, (3, 3))
mat_b = np.random.randint(1, 11, (3, 3))
diff = mat_a - mat_b
print("\nМатриця A:\n", mat_a)
print("Матриця B:\n", mat_b)
print("Різниця A - B:\n", diff)

# 9. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.
mat_c = np.random.randint(-5, 6, (3, 3))
mat_d = np.random.randint(-5, 6, (3, 3))
product2 = np.dot(mat_c, mat_d)
print("\nМатриця C:\n", mat_c)
print("Матриця D:\n", mat_d)
print("Добуток C * D:\n", product2)

# 10. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
mat_e = np.random.uniform(-1, 1, (3, 3))
row_sums = np.sum(mat_e, axis=1).reshape(-1, 1)
print("\nМатриця 3x3 з числами [-1, 1]:\n", mat_e)
print("Сума елементів кожного рядка (вектор-стовпчик):\n", row_sums)
```
